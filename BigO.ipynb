# Functions and Complexity

```python
def vector_dot_product(v1, v2): 
    """
    Computes the dot product of two vectors.
    """
    return sum([v1[i] * v2[i] for i in range(len(v1))])

# the complexity of the code (O notation)
# O(n) = n
```

```python
def vector_cross_product(v1, v2): 
    """
    Computes the cross product of two 3D vectors.
    """
    if len(v1) != 3 or len(v2) != 3:
        raise ValueError("Both vectors must be 3D vectors.")
    else:
        return [v1[1] * v2[2] - v1[2] * v2[1], 
                v1[2] * v2[0] - v1[0] * v2[2], 
                v1[0] * v2[1] - v1[1] * v2[0]]

# the complexity of the code (O notation)
# O(n) = 1 # constant, since it's a fixed 3D calculation
```

```python
def matrix_addition(A, B): 
    """
    Adds two matrices of the same size.
    """
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^2 # complexity for n x n matrices
```

```python
def matrix_subtraction(A, B):
    """
    Subtracts matrix B from matrix A, element-wise.
    """
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^2 # complexity for n x n matrices
```

```python
def matrix_product(A, B): 
    """
    Multiplies two matrices.
    """
    return [[sum([A[i][k] * B[k][j] for k in range(len(A[0]))]) for j in range(len(B[0]))] for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^3 # complexity for n x n matrices
```

```python
def matrix_vector_product(A, v): 
    """
    Multiplies a matrix by a vector.
    """
    return [sum([A[i][j] * v[j] for j in range(len(v))]) for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^2 # complexity for n x n matrix and vector of size n
```

```python
import time
import numpy as np

# 1. vector_dot_product(v1, v2): Computes the dot product of two vectors.
def vector_dot_product(v1, v2):
    result = 0
    for i in range(len(v1)):
        result += v1[i] * v2[i]
    return result

# 2. vector_cross_product(v1, v2): Computes the cross product of two 3D vectors.
def vector_cross_product(v1, v2):
    return [
        v1[1] * v2[2] - v1[2] * v2[1],
        v1[2] * v2[0] - v1[0] * v2[2],
        v1[0] * v2[1] - v1[1] * v2[0]
    ]

# 3. matrix_addition(A, B): Adds two matrices of the same size (m x n).
def matrix_addition(A, B):
    result = [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
    return result

# 4. matrix_subtraction(A, B): Subtracts matrix B from matrix A, element-wise (m x n).
def matrix_subtraction(A, B):
    result = [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
    return result

# 5. matrix_product(A, B): Multiplies two matrices (A is m x n, B is n x p).
def matrix_product(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):           
        for j in range(len(B[0])):    
            for k in range(len(B)):   
                result[i][j] += A[i][k] * B[k][j]
    return result

# 6. matrix_vector_product(A, v): Multiplies a matrix A (m x n) by a vector v (n).
def matrix_vector_product(A, v):
    result = [0 for _ in range(len(A))]
    for i in range(len(A)):           
        for j in range(len(v)):       
            result[i] += A[i][j] * v[j]
    return result

# Timing comparison with NumPy
def time_comparison():
    v1 = [1, 2, 3]
    v2 = [4, 5, 6]
    A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
    B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]
    
    start_time = time.time()
    vector_dot_product(v1, v2)
    custom_dot_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    vector_cross_product(v1, v2)
    custom_cross_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    matrix_addition(A, B)
    custom_addition_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    matrix_subtraction(A, B)
    custom_subtraction_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    matrix_product(A, B)
    custom_product_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    matrix_vector_product(A, v1)
    custom_mv_product_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    # NumPy implementations for comparison
    np_v1 = np.array(v1)
    np_v2 = np.array(v2)
    np_A = np.array(A)
    np_B = np.array(B)
    
    start_time = time.time()
    np.dot(np_v1, np_v2)
    numpy_dot_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    np.cross(np_v1, np_v2)
    numpy_cross_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    np.add(np_A, np_B)
    numpy_addition_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    np.subtract(np_A, np_B)
    numpy_subtraction_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    np.matmul(np_A, np_B)
    numpy_product_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds
    
    start_time = time.time()
    np.dot(np_A, np_v1)
    numpy_mv_product_time = (time.time() - start_time) * 1e9  # Convert to nanoseconds

    # Display timing results in nanoseconds
    print("Custom vs NumPy Timing (in nanoseconds):")
    print(f"Dot Product: Custom = {custom_dot_time:.3f} ns, NumPy = {numpy_dot_time:.3f} ns, Difference = {custom_dot_time - numpy_dot_time:.3f} ns")
    print(f"Cross Product: Custom = {custom_cross_time:.3f} ns, NumPy = {numpy_cross_time:.3f} ns, Difference = {custom_cross_time - numpy_cross_time:.3f} ns")
    print(f"Matrix Addition: Custom = {custom_addition_time:.3f} ns, NumPy = {numpy_addition_time:.3f} ns, Difference = {custom_addition_time - numpy_addition_time:.3f} ns")
    print(f"Matrix Subtraction: Custom = {custom_subtraction_time:.3f} ns, NumPy = {numpy_subtraction_time:.3f} ns, Difference = {custom_subtraction_time - numpy_subtraction_time:.3f} ns")
    print(f"Matrix Product: Custom = {custom_product_time:.3f} ns, NumPy = {numpy_product_time:.3f} ns, Difference = {custom_product_time - numpy_product_time:.3f} ns")
    print(f"Matrix-Vector Product: Custom = {custom_mv_product_time:.3f} ns, NumPy = {numpy_mv_product_time:.3f} ns, Difference = {custom_mv_product_time - numpy_mv_product_time:.3f} ns")

# Run the time comparison
time_comparison()
```

# Use case: Logistic Regression algorithm

```python
import matplotlib.pyplot as plt

# Number of data points per class
n_data = 100

# Generate class 0: centered around (-2, -2) with Gaussian noise
class_0 = np.random.randn(n_data // 2, 2) + np.array([0, 0])

# Generate class 1: centered around (2, 2) with Gaussian noise
class_1 = np.random.randn(n_data // 2, 2) + np.array([2, 2])

# Combine the data and create labels
X = np.vstack((class_0, class_1))  # Combine the two classes into one dataset
y = np.hstack((np.zeros(n_data // 2), np.ones(n_data // 2)))  # 0 for class 0, 1 for class 1

# Visualize the generated data
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='blue', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='red', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Generated 2D Data for Logistic Regression')
plt.show()
```

```python
%%time

# In this cell we implement Logistic Regression algorithm without numpy:
import math

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Binary cross-entropy loss
def binary_cross_entropy(y, y_hat):
    if y_hat == 0:
        y_hat = 1e-15  # To prevent log(0)
    elif y_hat == 1:
        y_hat = 1 - 1e-15  # To prevent log(1)
    return -y * math.log(y_hat) - (1 - y) * math.log(1 - y_hat)

# Logistic Regression without NumPy
def logistic_regression(X, y, learning_rate=0.01, n_epochs=100):
    n_data = len(y)  # Number of training examples
    w0, w1, w2 = random.random(), random.random(), random.random()  # Initialize weights
    
    for epoch in range(n_epochs):
        total_loss = 0
        
        # Gradient accumulators
        dw0, dw1, dw2 = 0, 0, 0
        # Iterate over each training example
        for i in range(n_data):
            x1, x2 = X[i]
            z = w0 + w1 * x1 + w2 * x2
            y_hat = sigmoid(z)

            # Compute loss and accumulate
            loss = binary_cross_entropy(y[i], y_hat)
            total_loss += loss
                        
            # Gradients for the weights
            dw0 += (y_hat - y[i])
            dw1 += (y_hat - y[i]) * x1
            dw2 += (y_hat - y[i]) * x2

        
        # Update the weights using gradient descent
        w0 -= learning_rate * dw0 / n_data
        w1 -= learning_rate * dw1 / n_data
        w2 -= learning_rate * dw2 / n_data
        
        # Average loss for the epoch
        avg_loss = total_loss / n_data
        
        # Compute accuracy
        correct_predictions = 0
        for i in range(n_data):
            z = w0 + w1 * X[i][0] + w2 * X[i][1]
            y_hat = sigmoid(z)
            y_pred = 1 if y_hat > 0.5 else 0
            if y_pred == y[i]:
                correct_predictions += 1
        
        accuracy = correct_predictions / n_data
        
        # Print accuracy and loss for the current epoch
        print(f'Epoch {epoch + 1}/{n_epochs} - Accuracy: {accuracy:.4f}, Loss: {avg_loss:.4f}')
        
        # Early stopping criterion (if the loss stops decreasing)
        if epoch > 0 and abs(avg_loss - prev_loss) < 1e-10:
            break
        
        prev_loss = avg_loss

# Train the logistic regression model
logistic_regression(X, y, learning_rate=0.01, n_epochs=1000)
```
