# Functions and Complexity

```python
def vector_dot_product(v1, v2): 
    """
    Computes the dot product of two vectors.
    """
    return sum([v1[i] * v2[i] for i in range(len(v1))])

# the complexity of the code (O notation)
# O(n) = n
```

```python
def vector_cross_product(v1, v2): 
    """
    Computes the cross product of two 3D vectors.
    """
    if len(v1) != 3 or len(v2) != 3:
        raise ValueError("Both vectors must be 3D vectors.")
    else:
        return [v1[1] * v2[2] - v1[2] * v2[1], 
                v1[2] * v2[0] - v1[0] * v2[2], 
                v1[0] * v2[1] - v1[1] * v2[0]]

# the complexity of the code (O notation)
# O(n) = 1 # constant, since it's a fixed 3D calculation
```

```python
def matrix_addition(A, B): 
    """
    Adds two matrices of the same size.
    """
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^2 # complexity for n x n matrices
```

```python
def matrix_subtraction(A, B):
    """
    Subtracts matrix B from matrix A, element-wise.
    """
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^2 # complexity for n x n matrices
```

```python
def matrix_product(A, B): 
    """
    Multiplies two matrices.
    """
    return [[sum([A[i][k] * B[k][j] for k in range(len(A[0]))]) for j in range(len(B[0]))] for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^3 # complexity for n x n matrices
```

```python
def matrix_vector_product(A, v): 
    """
    Multiplies a matrix by a vector.
    """
    return [sum([A[i][j] * v[j] for j in range(len(v))]) for i in range(len(A))]

# the complexity of the code (O notation)
# O(n) = n^2 # complexity for n x n matrix and vector of size n
```
